#**üîñ Index operators loc[ ] & iloc[ ]**

**ÏõêÏÜå ÏÑ†ÌÉù**
* DataFrameObject.loc[rowÌñâÏù∏Îç±Ïä§, columnÏó¥Ïù¥Î¶Ñ]
* DataFrameObject.iloc[rowÌñâÎ≤àÌò∏, columnÏó¥Î≤àÌò∏]

**Î≤îÏúÑ Ïä¨ÎùºÏù¥Ïã± ÌôúÏö©**

* DataFrameObject.iloc[ÏãúÏûëÏù∏Îç±Ïä§:ÎÅùÏù∏Îç±Ïä§:Ïä¨ÎùºÏù¥Ïã±Í∞ÑÍ≤©]
* The first colon : indicates that we're selecting all rows.
* The second colon : indicates the step size, which is

```python
import pandas as pd

exam_data = {'Math':[90, 80, 70, 85], 'English':[98, 89, 95, 79],'Music':[85, 95, 100, 100], 'Gym':[100,90,90, 88]}
df=pd.DataFrame(exam_data, index=['Simon','Emma','Mary','Louis'])
print(df)
print('\n')
df1=df.iloc[::2]
print(df1)
print('\n')
df2=df.iloc[0:3]
print(df2)
print('\n')
df3=df.iloc[:-1]
print(df3)
```

####<font color = 'red'> üëÄüêæ **Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Î≥ÄÌôò Ïó∞Ïäµ Í≥ÑÏÜç Converting to basic data frame from dictionary data** 1 ‚§µÔ∏è

```python
import pandas as pd
dict_data={'c0':[100,200,300],
           'c1':[400,500,600],
           'c2':[700,800,900],
           'c3':[1000,1100,1200],
           'c4':[1300,1400,1500]}
df=pd.DataFrame(dict_data)
print(df)
print('\n')

df=pd.DataFrame(dict_data,index=['r0','r1','r2'])
print(df)
print('\n')

new_index = ['r0','r1','r2','r3','r4']
ndf = df.reindex(new_index)
print(ndf)
print('\n') #not a number NAN

new_index=['r0','r1','r2','r3','r4']
ndf2=df.reindex(new_index,fill_value=0)
print(ndf2)
print('\n')

ndf2=df.reset_index()
print(ndf2)
print('\n')

#ndf2.set_index('index',inplace=True)
#print(df)
#df=pd.DataFrame(dict_data,index=['r0','r1','r2'])
#print(df)
#print('\n')
ndf3=df.sort_index()
print(ndf3)
print('\n')

ndf4=df.sort_index(ascending=False)
print(ndf4)
print('\n')

ndf5=df.sort_values(by='c1',ascending=False)
print(ndf5)
ndf6=df.sort_values(by='c1',ascending=True)
print(ndf6)

```
